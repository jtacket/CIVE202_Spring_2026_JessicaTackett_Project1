```python
## Load data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

## Turn data into a dataframe
air_data = pd.DataFrame(air_data)
```

```python
# Run statistics for multiple groups (VOC, PM 2.5, PM 10.0)
concentration_summary = air_data.groupby(['sensor.name']).agg({'voc': ['mean','median']},{'pm2.5_atm':  ['mean','median'], 'pm10.0_atm': ['mean','median']},)

# Create a pivot table to show the top five locations with the highest mean and median in a clear and concise manner
pivot_stats = pd.pivot_table(
    air_data,
    index = "sensor.name",
    values = ["voc","pm2.5_atm", "pm10.0_atm"],
    aggfunc = ["mean", "median"])

# Prints the top five location for the pivot table
pivot_stats.head(5)
```

Task #3: Does humidity and temperature have a noticeable effect on air quality?
    a. Please use the following categories for humidity:
        i. Low humidity occurs at a relative humidity less than 50%, high humidity occurs at a relative humidity between 50 and 80%, very high humidity occurs at a relative humidity above 80%

```python
# Define a function that assigns the temperatures to different categories

def assign_temperature(temperature): 
    if temperature < 32: 
        return "Below Freezing" 
    elif 32<= temperature <= 50: 
        return "Cool" 
    elif 51 <= temperature <= 70: 
        return "Warm" 
    else: 
        return "Hot" 

# Define a function that assigns the humidity to different categories

def assign_humidity(humidity): 
    if humidity < 50: 
        return "Low" 
    elif 51<= humidity <= 80: 
        return "High" 
    else: 
        return "Very High"  

# Create a new column for the humidity and temperature level

air_data["temperature_level"] = air_data["temperature"].apply(assign_temperature)
air_data["humidity_level"] = air_data["humidity"].apply(assign_humidity) 

# Group the data summary and run the statistics for the max, min, mean, and median
 
data_summary = air_data.groupby(['humidity_level', 'temperature']).agg({ 

    'pm2.5_atm': ['max','min','mean','median'], 

    'pm10.0_atm': ['max','min','mean','median'],
    'voc': ['max','min','mean','median']}) 
 

# Print a Pivot table to show the summary in a clear way

pivot_stats = pd.pivot_table( 

    air_data, 

    index = ['humidity_level', 'temperature_level'], 

    values = ['pm2.5_atm','pm10.0_atm','voc'], 

    aggfunc = ['max','min','mean','median']) 

pivot_stats 
```

Task #5 BONUS: Does the sensor altitude have an impact on the air quality values?

```python
# Run statistics for the polutants correlating to the sensor altitude
altitude_summary = air_data.groupby(['sensor.altitude']).agg({'voc': ['max','min','mean','median']},{'pm2.5_atm':  ['max','min','mean','median'], 'pm10.0_atm': ['max','min','mean','median']},)

# Last, I am going to make a table showing the correlation of the data to see if sensor altitude has an impact on air quality values
correlation_data = air_data[["sensor.altitude","voc","pm10.0_atm","pm2.5_atm"]].corr()

print(correlation_data)

# 1.0 = very positive relationship, 0.5 = moderate positive relationship, 0 = no relationship,
#-0.5 = moderate negative relationship, -1.0 = very negative relationship
```
